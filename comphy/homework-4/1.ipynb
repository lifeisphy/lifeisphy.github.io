```python
from matplotlib import pyplot as plt
from math import *
import numpy as np
import scipy
import sympy
# 该文件用作检验比对
```

```python
N=10

def mod(idx):
    if(idx<0):
        return idx+N
    elif(idx>=N):
        return idx-N
    else:
        return idx

def delta(i,j):
    return 1 if mod(i) == mod(j) else 0

def f(i,j):
    return delta(i-1,j)+delta(i+1,j)-2*delta(i,j)
mat = [[f(i,j) for i in range(N)] for j in range(N)]
arr=np.matrix(mat,dtype='float')
print(arr)
```

```python
np.linalg.eig(arr)
```
